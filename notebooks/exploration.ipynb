```python
from pathlib import Path
import geopandas as gpd

path = Path("../data/interim/crashes_with_segments.geojson")
gdf = gpd.read_file(path)

print("SHAPE")
print(gdf.shape)

print("\nCOLUMNS")
print(list(gdf.columns))

print("\nHEAD")
print(gdf[["crash_uid", "collision_id", "segment_id", "distance_to_segment"]].head(10))

print("\nDUPLICATE CHECK")
print("unique crash_uid:", gdf["crash_uid"].nunique())
print("duplicate crash_uid rows:", gdf.duplicated(subset=["crash_uid"]).sum())
print("null segment_id:", gdf["segment_id"].isna().sum())

print("\nDISTANCE SUMMARY")
print(gdf["distance_to_segment"].describe())

print("\nSEGMENT CHECK")
print("unique segment_id:", gdf["segment_id"].astype(str).nunique())
print(gdf["segment_id"].astype(str).value_counts().head(10))
```

```python
import pandas as pd

df = pd.read_csv("../data/processed/segment_daily_crashes.csv")

print("SHAPE")
print(df.shape)

print("\nCOLUMNS")
print(list(df.columns))

print("\nHEAD")
print(df.head(10))

print("\nDTYPES")
print(df.dtypes)

print("\nDATE RANGE")
d = pd.to_datetime(df["date"], errors="coerce")
print("min:", d.min())
print("max:", d.max())
print("nulls after parse:", d.isna().sum())

print("\nSEGMENT CHECK")
print("unique segment_id:", df["segment_id"].astype(str).nunique())
print(df["segment_id"].astype(str).value_counts().head(10))

print("\nCRASH COUNT CHECK")
print(df["crash_count"].value_counts().head(20))

print("\nCRASH COUNT SUMMARY")
print(df["crash_count"].describe())
```

```python
import pandas as pd

seg = pd.read_csv("../data/processed/segment_features.csv")
weather = pd.read_csv("../data/raw/nyc_weather_2018_2024.csv")

print("SEGMENT FEATURES SHAPE")
print(seg.shape)
print("\nSEGMENT FEATURES HEAD")
print(seg.head(10))
print("\nSEGMENT FEATURES DTYPES")
print(seg.dtypes)
print("\nROAD TYPE COUNTS")
print(seg["road_type"].value_counts().head(15))
print("\nROAD TYPE UNIQUE EXAMPLES")
print(seg["road_type"].unique()[:15])

print("\n" + "="*80 + "\n")

print("WEATHER SHAPE")
print(weather.shape)
print("\nWEATHER HEAD")
print(weather.head(10))
print("\nWEATHER DTYPES")
print(weather.dtypes)
print("\nWEATHER DATE RANGE")
wd = pd.to_datetime(weather["date"], errors="coerce")
print("min:", wd.min())
print("max:", wd.max())
print("nulls after parse:", wd.isna().sum())
print("\nWEATHER NULLS")
print(weather.isna().sum())
```
